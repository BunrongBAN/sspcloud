## Slicing python  
### method = lst[first_position,end_position,step]


```python
lst = [1,2,3,4,7,9]
print(lst[4:-1])
print(lst[2:5:1])
```

```python
x1 = np.arange(10)
print(x1[1:6:2],x1)
print(x1[0:10:2], 'ou', x1[::2]) 
print(x1[1:len(x1)])
print(x1[0:5], 'ou', x1[:5])
```
